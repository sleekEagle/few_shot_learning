```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

wv_file='/home/lnw8px/models/HAR_research/3 methods/PAMAP2_WV_5_classes_tune_lambda.txt'
#center_file='/home/lnw8px/models/HAR_research/3 methods/PAMAP2_tune_LAMBDA_bs4_discount0.8_emb128.txt'
center_file='/home/lnw8px/models/HAR_research/3 methods/PAMAP2_tune_LAMBDA_bs4_discount0.8_emb128.txt'
softmax_file='/home/lnw8px/models/HAR_research/3 methods/PAMAP2_center_modelcomplexity.txt'
```

```python
df=pd.read_csv(softmax_file,delimiter=';',header=None)
df.columns=['num_kernels','num_unseen_classes','train_acc','num_samples','knn_class_acc','knn_acc']
#df.columns=['segment_len','num_embeddings','train_acc','num_samples','knn_class_acc','knn_acc','unseen_classes']

def get_num(row):
    return float(row[0:-1])
df['knn_acc']=df['knn_acc'].apply(get_num)
```

```python
'''
plot knn_acc vs num samples for some LAMBDA for word2vec embedding
'''
acc_list=[]
num_samples_list=[]
for num_samples in range(2,10):
    acc_list.append(df[df['num_samples']==num_samples].groupby(['lambda']).mean().loc[5.000000e-04]['knn_acc'])
    num_samples_list.append(num_samples)
plt.scatter(num_samples_list,acc_list)
plt.xlabel('number of samples for knn train')
plt.ylabel('knn accuracy')
plt.title('For word2vec embedding, LAMBDA=0.0001,ALPHA=0.5')
```

```python
'''
plot knn_acc vs num samples for some LAMBDA for kinetic embedding
'''
acc_list=[]
num_samples_list=[]
for num_samples in range(2,10):
    d=df[df['num_samples']==num_samples]
    acc_list.append(d[(d['lambda']==0.01)&(d['alpha']==1)]['knn_acc'].iloc[0])
    num_samples_list.append(num_samples)
plt.scatter(num_samples_list,acc_list)
plt.xlabel('number of samples for knn train')
plt.ylabel('knn accuracy')
plt.title('For kinetic embedding, LAMBDA=0.0001,ALPHA=0.5')
```
